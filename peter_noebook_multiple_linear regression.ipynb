### Data cleaning

```python
#dropping the 'Location_1' column which has no values
pop.drop('Location_1',axis=1,inplace=True)

#finding the mean of House_density column and using the mean to fill the empty column
pop['_Household_Density_'].fillna(pop['_Household_Density_'].mean(),inplace=True)

#finding the mean of Population_density column and using the mean to fill the empty column
pop['Population_Density'].fillna(pop['Population_Density'].mean(),inplace=True)

#finding the mean of Gender_index column and using the mean to fill the empty column
pop['Gender_Index'].fillna(pop['Gender_Index'].mean(),inplace=True)

```

```python
### The next step is to divide the data into attributes and labels 
X=pop[['Male','Female','Households']]
y=pop['Population_Density']
```

### Implementing multiple linear regeression


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

### Training the algorithm

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#the three coefficients of our multiple linear regression
coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

### Making prediction

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
#being an numpy array we convert the array into list appending the values into a new list called y_pred_final and compare the predicted values against the test values

y_pred_final=[]
for i in y_pred:
    y_pred_final.append(int(i))
    
y_pred_final
```

```python
y_real=list(y_test)
```

```python
# convert the outcome into a dictionary/dataframe for ease of comparison
data={'Predicted':y_pred_final,'Test_values':y_real}
data_df=pd.DataFrame(data)
data_df
```

### Model evaluation

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
